```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.signal import welch
import matplotlib.pyplot as plt

# --- 全局专业绘图风格设置 ---
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['SimHei', 'Arial'], # 优先使用黑体，备选Arial
    'axes.unicode_minus': False,
    'figure.figsize': (16, 9),
    'axes.labelsize': 14,
    'axes.titlesize': 18,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'legend.fontsize': 13,
    'lines.linewidth': 2.0,
})

# ==============================================================================
#  模型定义与仿真计算部分 (与之前版本相同, 此处为简洁可折叠)
# ==============================================================================
class BrownianNoiseGenerator:
    """生成更真实的布朗噪声激励源"""
    def __init__(self, t_eval, scale=1.0, seed=42):
        np.random.seed(seed)
        n_steps = len(t_eval)
        dt = t_eval[1] - t_eval[0]
        white_noise = np.random.randn(n_steps) * np.sqrt(dt)
        self.brownian_noise = np.cumsum(white_noise)
        self.brownian_noise -= np.mean(self.brownian_noise)
        self.brownian_noise = (self.brownian_noise / np.std(self.brownian_noise)) * scale
        self.t_eval = t_eval

    def get_force(self, t):
        return np.interp(t, self.t_eval, self.brownian_noise)

class VibrationModel:
    """封装振动模型，用于仿真松动和老化"""
    def __init__(self, m, excitation_generator):
        self.m = m
        self.excitation = excitation_generator
        self.y0 = [0, 0]

    def _ode_system(self, t, y, omega_n_func, zeta_func):
        x, dx_dt = y
        force = self.excitation.get_force(t)
        omega_n = omega_n_func(t)
        zeta = zeta_func(t)
        d2x_dt2 = (force / self.m) - 2 * zeta * omega_n * dx_dt - omega_n**2 * x
        return [dx_dt, d2x_dt2]

    def simulate(self, t_span, t_eval, omega_n_func, zeta_func):
        return solve_ivp(
            self._ode_system, t_span, self.y0,
            args=(omega_n_func, zeta_func), dense_output=True, t_eval=t_eval
        )

class FluidModel:
    """封装流体模型，用于仿真渗漏和变形"""
    def __init__(self, V0, beta, rho, excitation_generator):
        self.V0, self.beta, self.rho = V0, beta, rho
        self.excitation = excitation_generator
        self.P0 = [1e6]

    def _ode_system(self, t, P, Cd_out, A_out, Cd_leak, A_leak):
        P_val = max(P[0], 0)
        q_in = max(0, 0.01 + self.excitation.get_force(t) * 0.001)
        q_out = Cd_out * A_out * np.sqrt(2 * P_val / self.rho)
        q_leak = Cd_leak * A_leak * np.sqrt(2 * P_val / self.rho)
        dP_dt = (self.beta / self.V0) * (q_in - q_out - q_leak)
        return [dP_dt]

    def simulate(self, t_span, t_eval, Cd_out, A_out, Cd_leak, A_leak):
        return solve_ivp(
            self._ode_system, t_span, self.P0,
            args=(Cd_out, A_out, Cd_leak, A_leak), dense_output=True, t_eval=t_eval
        )

# ==============================================================================
#  全新、独立的绘图函数
# ==============================================================================

def plot_excitation(generator):
    """图 1: 绘制激励源的时域与频域特性"""
    fig, axs = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle('仿真激励源特性分析', fontsize=22)
    
    # 时域图
    axs[0].plot(generator.t_eval, generator.brownian_noise, color='black', lw=1.5)
    axs[0].set_title('激励源时域波形')
    axs[0].set_xlabel('时间 (s)')
    axs[0].set_ylabel('激励幅值 (N or m³/s)')
    axs[0].grid(True, linestyle=':')

    # 频域图
    fs = 1 / (generator.t_eval[1] - generator.t_eval[0])
    f, Pxx = welch(generator.brownian_noise, fs, nperseg=1024)
    axs[1].loglog(f, Pxx, color='black', lw=1.5)
    axs[1].set_title('激励源功率谱密度 (PSD)')
    axs[1].set_xlabel('频率 (Hz)')
    axs[1].set_ylabel('功率谱密度')
    axs[1].grid(True, which='both', linestyle=':')
    
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

def plot_vibration_analysis(sol_healthy, sol_loose, sol_aged):
    """图 2: 振动分析 3x3 网格图 (行:工况, 列:分析方法)"""
    fig, axs = plt.subplots(3, 3, figsize=(20, 15), sharex='col') # 共享列的x轴
    fig.suptitle('振动故障诊断分析', fontsize=24)
    
    solutions = {
        '健康状态': (sol_healthy, 'C0'),
        '渐进式松动': (sol_loose, 'C1'),
        '渐进式老化': (sol_aged, 'C2')
    }
    
    for i, (state, (sol, color)) in enumerate(solutions.items()):
        t, x, v = sol.t, sol.y[0], sol.y[1]
        
        # --- 列 1: 时域分析 ---
        axs[i, 0].plot(t, x, color=color)
        axs[i, 0].set_ylabel('位移 (m)')
        axs[i, 0].set_title(f'{state} - 时域响应', fontsize=16)
        axs[i, 0].grid(True, linestyle=':')

        # --- 列 2: 频域分析 ---
        fs = 1.0 / (t[1] - t[0])
        f, Pxx = welch(x, fs, nperseg=2048)
        axs[i, 1].semilogy(f, Pxx, color=color)
        axs[i, 1].set_title(f'{state} - 功率谱密度', fontsize=16)
        axs[i, 1].set_ylabel('功率谱')
        axs[i, 1].set_xlim(0, 5) # 聚焦于我们关心的频段
        axs[i, 1].grid(True, which='both', linestyle=':')
        
        # --- 列 3: 相图分析 ---
        axs[i, 2].plot(x, v, color=color, lw=0.5, alpha=0.7)
        axs[i, 2].set_title(f'{state} - 相图', fontsize=16)
        axs[i, 2].set_xlabel('位移 (m)')
        axs[i, 2].set_ylabel('速度 (m/s)')
        axs[i, 2].grid(True, linestyle=':')

    # 设置最底行的x轴标签
    for j in range(3):
        axs[2, j].set_xlabel('时间 (s)' if j==0 else ('频率 (Hz)' if j==1 else '位移 (m)'))

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

def plot_fluid_analysis(sol_healthy, sol_faulty):
    """图 3: 流体分析 1x2 对比图"""
    fig, axs = plt.subplots(1, 2, figsize=(18, 7))
    fig.suptitle('流体渗漏/变形故障诊断分析', fontsize=22)

    # --- 图 1: 时域压力对比 ---
    axs[0].plot(sol_healthy.t, sol_healthy.y[0] / 1e6, label='健康状态 (无泄漏)', lw=2)
    axs[0].plot(sol_faulty.t, sol_faulty.y[0] / 1e6, label='渗漏/变形故障', lw=2, alpha=0.8)
    axs[0].set_title('时域压力对比')
    axs[0].set_xlabel('时间 (s)')
    axs[0].set_ylabel('压力 (MPa)')
    axs[0].legend()
    axs[0].grid(True, linestyle=':')

    # --- 图 2: 压力概率密度分布 ---
    axs[1].hist(sol_healthy.y[0] / 1e6, bins=50, density=True, label='健康状态', alpha=0.75)
    axs[1].hist(sol_faulty.y[0] / 1e6, bins=50, density=True, label='渗漏/变形故障', alpha=0.75)
    axs[1].set_title('压力概率密度分布')
    axs[1].set_xlabel('压力 (MPa)')
    axs[1].set_ylabel('概率密度')
    axs[1].legend()
    axs[1].grid(True, linestyle=':')

    plt.tight_layout(rect=[0, 0, 1, 0.94])
    plt.show()

# ==============================================================================
#  主执行程序
# ==============================================================================
if __name__ == '__main__':
    # --- 1. 初始化和仿真计算 ---
    t_span = [0, 100]
    t_eval = np.linspace(t_span[0], t_span[1], 20000) # 增加点数以获得更平滑的谱

    # 初始化激励源
    excitation_gen = BrownianNoiseGenerator(t_eval, scale=5.0)

    # 振动模型仿真
    vib_model = VibrationModel(m=1.0, excitation_generator=excitation_gen)
    omega_n_healthy, zeta_healthy = 2 * np.pi * 3.0, 0.1
    def omega_n_loose_func(t): return np.interp(t, [0, 50], [2*np.pi*3.0, 2*np.pi*2.0])
    def omega_n_aged_func(t): return np.interp(t, [0, 50], [2*np.pi*3.0, 2*np.pi*3.2])
    def zeta_aged_func(t): return np.interp(t, [0, 50], [0.1, 0.02])

    sol_vib_healthy = vib_model.simulate(t_span, t_eval, lambda t: omega_n_healthy, lambda t: zeta_healthy)
    sol_vib_loose = vib_model.simulate(t_span, t_eval, omega_n_loose_func, lambda t: zeta_healthy)
    sol_vib_aged = vib_model.simulate(t_span, t_eval, omega_n_aged_func, zeta_aged_func)

    # 流体模型仿真
    fluid_model = FluidModel(V0=0.01, beta=2.2e9, rho=1000, excitation_generator=excitation_gen)
    A_out, A_leak = 1e-4, 0.2e-4
    sol_fluid_healthy = fluid_model.simulate(t_span, t_eval, 0.6, A_out, 0.6, 0.0)
    sol_fluid_faulty = fluid_model.simulate(t_span, t_eval, 0.6, A_out, 0.6, A_leak)

    # --- 2. 调用独立的绘图函数 ---
    print("正在生成图 1: 激励源特性...")
    plot_excitation(excitation_gen)
    
    print("\n正在生成图 2: 振动故障分析...")
    plot_vibration_analysis(sol_vib_healthy, sol_vib_loose, sol_vib_aged)

    print("\n正在生成图 3: 流体故障分析...")
    plot_fluid_analysis(sol_fluid_healthy, sol_fluid_faulty)
```
